### 3.2 Data cleansing

```python
#cleansing data : lower case all text, remove punctuation
def preprocessing(text):
    def strip_html(text):
        soup = BeautifulSoup(text, "html.parser")
        return soup.get_text()
    def remove_between_square_brackets(text):
        return re.sub('-\[[^]]*\]', '', text)
    def denoise_text(text):
        text = strip_html(text)
        text = remove_between_square_brackets(text)
        return text
    def remove_punctuation(text):
        return re.sub(r'[^\w\s]', '', text)
    def remove_non_ascii(text):
        return unicodedata.normalize('NFKD', text).encode('ascii', 'ignore').decode('utf-8', 'ignore')
    
    text = strip_html(text)
    text = remove_between_square_brackets(text)
    text = remove_punctuation(text)
    text = remove_non_ascii(text)
    
    return text.lower()
```

```python
# apply function, will implement on wordcloud
data['text'] = data['text'].progress_apply(preprocessing)
```

### 3.3 Stopwords removal
Remove stop words that won't give any insight, or save the words that will give impact.

```python
stop = stopwords.words('english')
data['text'] = data['text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
data['text'].head(10)
```

### 3.4 Frequent words removal
Remove the words that frequently shown in data.

```python
# find frequent words
freq = pd.Series(' '.join(data['text']).split()).value_counts()[:10]
freq
```

```python
# removing frequent words
freq = list(freq.index)
data['text'] = data['text'].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
data['text'].head()
```

### 3.5 Rare words removal
Remove the words that infrequently shown in data.

```python
# find the rare words
rare_word = pd.Series(' '.join(data['text']).split()).value_counts()[-10:]
rare_word
```

```python
# removing rare words
rare_word = list(rare_word.index)
data['text'] = data['text'].apply(lambda x: " ".join(x for x in x.split() if x not in freq))
data['text'].head()
```

### 3.6 Spelling correction
Correct spelling in english.

```python
data['text'][:5].apply(lambda x: str(TextBlob(x).correct()))
```

### 3.7 Tokenization
Convert paragraph or sentence into word.

```python
data['token'] = data['text']
data['token'] = data.apply(lambda row: nltk.word_tokenize(row['text']), axis=1)
data['token'].head()
```

### 3.8 Stemming
Reduces word-forms to (pseudo)stems.

```python
from nltk.stem import PorterStemmer
```

```python
st = PorterStemmer()
data['text'] = data['text'][:].apply(lambda x: " ".join([st.stem(word) for word in x.split()]))
```

### 3.9 Lemmatization
Reduces the word-forms to linguistically valid lemmas.

```python
data['text'] = data['text'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()]))
data['text'].head()
```

```python
# select sentiment that has confidence more than 0.5
conf = df[df['airline_sentiment_confidence']>0.5]

#count sentiment airline
sentiment = conf.groupby(['airline_sentiment'])['airline_sentiment'].count()
sentiment

#make dataframe that will be created a chart 
chart = pd.DataFrame(sentiment)
chart['tweet_count'] = chart['airline_sentiment']
chart.drop(['airline_sentiment'], axis = 1, inplace = True )
chart = chart.reset_index()

#make a proportion of the number of tweet of each sentiment 
chart['percent'] = chart['tweet_count']/chart['tweet_count'].sum()
chart
```

### 4.3 Find the most negative reason

##### Negative reason from all airlines

```python
#count negative reason
negative = conf.groupby(['negativereason'])['negativereason'].count()
negative

#make dataframe that will be created a chart 
reason = pd.DataFrame(negative)
reason['tweet_count'] = reason['negativereason']
reason.drop(['negativereason'], axis = 1, inplace = True )
reason = reason.reset_index()

#make a proportion of the number of tweet of each sentiment 
reason['percent'] = reason['tweet_count']/reason['tweet_count'].sum()
top5=reason.sort_values('percent', ascending=False).head()
top5
```

```python
#bar plot visualization 
plt.figure(figsize=(10,5))
ax = sns.barplot(x= top5['negativereason'], y='tweet_count', data=reason)
ax.set_title('Barplot Negative Reason')
```

From the above graph, we can see that the most reason of negative sentiment is customer service issue (the other reasons can be seen on the graph above).

##### Negative reason for each airlines

```python
# NR_Count = dict(df['negativereason'].value_counts(sort=False))
def count(Airline):
    if Airline=='All':
        df_new=df
    else:
        df_new=conf[conf['airline']==Airline]
    count=dict(df_new['negativereason'].value_counts())
    neg_reason=list(conf['negativereason'].unique())
    neg_reason=[x for x in neg_reason if str(x) != 'nan']
    df_reason=pd.DataFrame({'Reasons':neg_reason})
    df_reason['count']=df_reason['Reasons'].apply(lambda x: count[x])
    return df_reason

def plot_reason(Airline):
    df_count=count(Airline)
    ct=df_count['count']
    Index = range(1,(len(df_count)+1))
    plt.bar(Index,ct)
    plt.xticks(Index,df_count['Reasons'],rotation=90)
    plt.ylabel('Count')
    plt.xlabel('Reason')
    plt.title('Count of Reasons for '+Airline)
```

```python
# bar plot visualization for each airlines
airline_names = conf['airline'].unique()
plt.figure(1,figsize=(18,30))
for i in range(0,6):
    plt.subplot(3,3,i+1)
    plot_reason(airline_names[i])
    plt.xlabel('')
    plt.ylim(0,800)
plt.show()
```

From the above graph, we can see that the most reason of negative sentiment for each airline is customer service issue (except Delta airline which the most reason is late flight). These graph of negative reasons can be reference for each airlines to prioritize which issues should they fix first.

### 4.5 Word Cloud
Word cloud uses to see which word is the most importance and use on the tweet.

```python
#define each sentiment class into variabels
negative = df.loc[df['airline_sentiment'] == "negative"]
neutral  = df.loc[df['airline_sentiment'] == "neutral"]
positive = df.loc[df['airline_sentiment'] == "positive"]
```

```python
#merge all the sentiment class tweets into one paragraph
desc_negative = "".join(negative.text.values)
desc_neutral = "".join(neutral.text.values)
desc_positive = "".join(positive.text.values)
alls = desc_negative+desc_neutral+desc_positive
```

```python
#function to make a wordcloud for each sentiment class
def create_word_cloud(string):
    cloud = WordCloud(background_color = "white", max_words = 200, stopwords = set(STOPWORDS)).generate(string)
    plt.imshow(cloud, interpolation='nearest')
    plt.axis('off')
    plt.show()
```

```python
# all sentiment class
create_word_cloud(alls)
```

```python
# negative class 
create_word_cloud(desc_negative)
```

From the word cloud above, we can see that the most used word is **delay**.

```python
# neutral class
create_word_cloud(desc_neutral)
```

From the word cloud above, we can see that the most used word is **flight**.

```python
# positive class
create_word_cloud(desc_positive)
```

From the above word cloud, we can see that the most used word is **great**.
